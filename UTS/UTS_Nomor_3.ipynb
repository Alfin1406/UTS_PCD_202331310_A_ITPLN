```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
#202331310 - Muhammad Alfin Hariansyah
```

```python
# Membaca gambar backlight dan mengonversinya ke RGB
gambar_asli = cv2.imread("foto.jpg")
gambar_rgb = cv2.cvtColor(gambar_asli, cv2.COLOR_BGR2RGB)
#202331310 - Muhammad Alfin Hariansyah
```

```python
# Konversi gambar ke grayscale
gambar_gray = cv2.cvtColor(gambar_asli, cv2.COLOR_BGR2GRAY)
#202331310 - Muhammad Alfin Hariansyah
```

```python
# Fungsi untuk meningkatkan kecerahan gambar grayscale
def tingkatkan_kecerahan(gray_img, nilai=50):
    return cv2.add(gray_img, nilai)

# Fungsi untuk meningkatkan kontras menggunakan histogram equalization
def tingkatkan_kontras(gray_img):
    return cv2.equalizeHist(gray_img)
#202331310 - Muhammad Alfin Hariansyah
```

```python
# Proses perbaikan citra
gray_cerah = tingkatkan_kecerahan(gambar_gray)
gray_kontras = tingkatkan_kontras(gambar_gray)
cerah_kontras = tingkatkan_kontras(gray_cerah)

# Membaca gambar info (lokasi/pengguna kamera)
info_gambar = cv2.imread("bukti_foto.jpg")
info_rgb = cv2.cvtColor(info_gambar, cv2.COLOR_BGR2RGB)

# Daftar judul dan gambar yang akan ditampilkan
judul_gambar = [
    'Tampilan Awal (RGB)',
    'Grayscale Asli',
    'Grayscale Setelah Dicerahkan',
    'Grayscale Setelah Ditingkatkan Kontras',
    'Grayscale Setelah Cerah + Kontras',
    'Informasi Gambar & Lokasi'
]

daftar_gambar = [gambar_rgb, gambar_gray, gray_cerah, gray_kontras, cerah_kontras, info_rgb]
warna_cmap = [None, 'gray', 'gray', 'gray', 'gray', None]  # cmap None untuk gambar RGB
#202331310 - Muhammad Alfin Hariansyah
```

```python
# Plot semua gambar secara vertikal
plt.figure(figsize=(6, 24))  # Ukuran vertikal besar agar muat semua gambar
for idx in range(len(daftar_gambar)):
    plt.subplot(len(daftar_gambar), 1, idx + 1)
    plt.imshow(daftar_gambar[idx], cmap=warna_cmap[idx])
    plt.title(judul_gambar[idx], fontsize=12)
    plt.axis('off')

plt.tight_layout()
plt.show()
#202331310 - Muhammad Alfin Hariansyah
```
